# LDA


```python
import gensim
from gensim.models.ldamodel import LdaModel
```

```python
text_corpus = gt_data
```

```python
from gensim.corpora.dictionary import Dictionary
from gensim import corpora
# Create a corpus from a list of cleaned sentences

dictionary = corpora.Dictionary(text_corpus)
common_corpus = [dictionary.doc2bow(x) for x in text_corpus]
```

```python
# common_corpus
len(common_corpus)
# common_corpus[10]
```

```python
lda = LdaModel(common_corpus, num_topics=85)
```

```python
lda.save('model.lda')
```

```python
new_model_lda = gensim.models.LdaModel.load('/content/model.lda')
```

```python
new_model_lda
```

```python
lda_vector = []

for text in text_corpus:
  text_vec = [dictionary.doc2bow(text)]
  text_topics = new_model_lda[text_vec]

  temp_topic = []
  for topic in text_topics:
    temp_topic.append(topic)

  lda_vector.append(temp_topic)
```

```python
# lda_vector
print(len(lda_vector))
```

```python
for i in range(100,106):
  print(lda_vector[i])
```

# Clustering 5-cluster

```python
import numpy as np
```

```python
clustering_data = np.zeros((2779,85),dtype=float)
for i in range(2779):
  for a1 in lda_vector[i]:
    for a,b in a1:
      # print(a)
      # print(b)
      clustering_data[i][a]=b
```

```python
clustering_data[200]
```

```python
from sklearn.cluster import KMeans
```

```python
cluster_model = KMeans(n_clusters=5)
clusters = cluster_model.fit_predict(clustering_data)
```

```python
clusters
```

```python
# cluster_model.cluster_centers_
```

```python
cluster_cen = cluster_model.cluster_centers_
```

# Fairlet balance factor

```python
# for each cluster [muslim, hindu, undetected]
fairlet_cluster_stat = [ [0,0] for _ in range(5)]
```

```python
for i in range(2779):
  if gt_reli[i] == 1:
    fairlet_cluster_stat[clusters[i]][0] += 1
  elif gt_reli[i] == 2:
    fairlet_cluster_stat[clusters[i]][1] += 1
```

```python
fairlet_cluster_stat
```

```python
balance_factor = 1;
for i in range(5):
  balance_factor = min(balance_factor,
                       min(fairlet_cluster_stat[i][0]/fairlet_cluster_stat[i][1],fairlet_cluster_stat[i][1]/fairlet_cluster_stat[i][0]))
```

```python
balance_factor
```

# fair-clustering evaluation, samira samadi

```python
dist_hindu = 0
dist_muslim = 0
```

```python
for i in range(2779):
  if gt_reli[i]==1:
    x1 = (np.linalg.norm(clustering_data[i] - cluster_cen[clusters[i]]))
    dist_muslim += x1
  elif gt_reli[i]==2:
    x1 = (np.linalg.norm(clustering_data[i] - cluster_cen[clusters[i]]))
    dist_hindu += x1
```

```python
hindu_num, muslim_num = 0,0
for i in range(len(gt_reli)):
  if gt_reli[i]==1:
    muslim_num +=1
  else :
    hindu_num+=1
```

```python
dist_hindu/hindu_num
```

```python
dist_muslim/muslim_num
```
