#LDA

```python
import gensim
from gensim.models.ldamodel import LdaModel
from gensim.corpora.dictionary import Dictionary
from gensim import corpora
text_corpus = gt_data
dictionary = corpora.Dictionary(text_corpus)
common_corpus = [dictionary.doc2bow(x) for x in text_corpus]
```

```python
def lda_creation(num_topic,n_word):
  lda = LdaModel(common_corpus, num_topics=num_topic,minimum_probability=0)
  lda_vector = []
  for text in text_corpus:
    text_vec = [dictionary.doc2bow(text)]
    text_topics = lda[text_vec]

    temp_topic = []
    for topic in text_topics:
      temp_topic.append(topic)

    lda_vector.append(temp_topic)

  x_tp = lda.show_topics(num_topics= num_topic, num_words=n_word,formatted=False)
  topics_words = [(tp[0], [wd[0] for wd in tp[1]]) for tp in x_tp]
  # for topic,words in topics_words:
  #   print(str(topic)+ "::"+ str(words))
  #   print()
  # # print()
  topic_wds = []
  for i in range(num_topic):
    tp_wd = list(w for w,_ in lda.show_topic(topicid = i,topn=n_word))
    temp_wds = []
    for x in tp_wd:
      x1 = int(x)
      temp_wds.append(dictionary.get((x1)))
    topic_wds.append(temp_wds)

  return lda_vector,topic_wds
```

# PCA and Reconstruction Error

```python
def re_error(n,x_data,topic_num):
  total_cost, hindu_cost, muslim_cost = 0,0,0

  #mean calculation
  mean_data = np.zeros((1,topic_num))
  for i in range(2020):
    mean_data += pca_data[i]
  mean_data /=2020
  for i in range(2020):
    for j in range(30):
      pca_data[i][j] -= mean_data[0][j]

  #pca calculation
  cov_mat = np.cov(np.transpose(pca_data))
  eig_vals, eig_vecs = np.linalg.eig(cov_mat)
  idx = eig_vals.argsort()[::-1]
  eig_vals = eig_vals[idx]
  eig_vecs = eig_vecs[:,idx]

  # reconstruction error calculation
  x=0
  for i in range(2020):
    x = 0
    z = np.zeros((1,topic_num))
    for j in range(n):
      z += (((np.dot(np.transpose(pca_data[i]),eig_vecs[j]))) * eig_vecs[j])

    x += (np.linalg.norm (z - pca_data[i]))

    total_cost += x
    if gt_reli[i]==1:
      muslim_cost += x
    else :
      hindu_cost += x
  # print(total_cost)

  total_cost /= 2020
  hindu_cost /= 1010
  muslim_cost /= 1010

# multiplier = np.dot(np.transpose(pca_data[0]),eig_vecs[0])
# recon_data = multiplier* eig_vecs[0]
# a = np.linalg.norm(recon_data - pca_data[0])



  return total_cost, hindu_cost, muslim_cost
```

```python
reconstruction_error_total, reconstruction_error_hindu, reconstruction_error_muslim = [], [], []

for i in range(1,20):
  re_t,re_h,re_m = re_error(i,pca_data,30)
  reconstruction_error_total.append(re_t)
  reconstruction_error_hindu.append(re_h)
  reconstruction_error_muslim.append(re_m)
  # print()
```

```python
def temp_pca_re_error(n,x_data,topic_num):
  total_cost, t1, t2 = 0, 0, 0

  pca_data = x_data
  #mean calculation
  mean_data = np.zeros((1,topic_num))
  for i in range(1010):
    mean_data += pca_data[i]
  mean_data /=1010
  for i in range(1010):
    for j in range(30):
      pca_data[i][j] -= mean_data[0][j]

  #pca calculation
  cov_mat = np.cov(np.transpose(pca_data))
  eig_vals, eig_vecs = np.linalg.eig(cov_mat)
  idx = eig_vals.argsort()[::-1]
  eig_vals = eig_vals[idx]
  eig_vecs = eig_vecs[:,idx]

  # reconstruction error calculation
  x=0
  for i in range(1010):
    x = 0
    z = np.zeros((1,topic_num))
    for j in range(n):
      z += (((np.dot(np.transpose(pca_data[i]),eig_vecs[j]))) * eig_vecs[j])

    x += (np.linalg.norm (z - pca_data[i]))

    total_cost += x
    if i%2 == 0:
      t1+=x
    else :
      t2+=x
  # print(total_cost)

  total_cost /= 1010
  t1 /= 505
  t2 /= 505


  return total_cost, t1,t2
```

```python
def temp_same_pca_re_error(n,x_data,topic_num):
  total_cost, t11, t12, t21, t22 = 0, 0, 0, 0, 0

  pca_data = x_data
  #mean calculation
  mean_data = np.zeros((1,topic_num))
  for i in range(1010):
    mean_data += pca_data[i]
  mean_data /=1010
  for i in range(1010):
    for j in range(30):
      pca_data[i][j] -= mean_data[0][j]

  #pca calculation
  cov_mat = np.cov(np.transpose(pca_data))
  eig_vals, eig_vecs = np.linalg.eig(cov_mat)
  idx = eig_vals.argsort()[::-1]
  eig_vals = eig_vals[idx]
  eig_vecs = eig_vecs[:,idx]

  # reconstruction error calculation
  x=0
  ok, flag = True,True
  for i in range(2020):
    x = 0
    z = np.zeros((1,topic_num))
    for j in range(n):
      z += (((np.dot(np.transpose(pca_data[i]),eig_vecs[j]))) * eig_vecs[j])

    x += (np.linalg.norm (z - pca_data[i]))

    total_cost += x
    if gt_reli[i]==2:
      if ok :
        t11 += x
        ok = False
      else :
        t12 += x
        ok = True
    else :
      if flag:
        t21 += x
        flag = False
      else:
        t22+= x
        flag = True



  # if gt_reli[i]==1:
  #     muslim_cost += x
  #   else :
  #     hindu_cost += x
  # print(total_cost)

  total_cost /= 2020
  t11 /= 505
  t12 /= 505
  t21 /= 505
  t22 /= 505


  return total_cost, t11, t12, t21, t22
```
